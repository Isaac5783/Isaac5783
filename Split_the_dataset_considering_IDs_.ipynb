```python
import shutil
```

```python
# Paths to the directories
source_directory = '/content/drive/MyDrive/dataset'
image_directory_T = '/content/drive/MyDrive/dataset/images/test'
image_directory_V = '/content/drive/MyDrive/dataset/images/val'
label_directory_T = '/content/drive/MyDrive/dataset/labels/test'
label_directory_V = '/content/drive/MyDrive/dataset/labels/val'
```

```python
image_filenames = os.listdir(source_directory)
```

```python
# Initialize dictionaries to hold the count of records for each patient
record_files = defaultdict(list)
patient_record_count = defaultdict(int)
```

```python
# Populate the record files and patient record counts
for filename in image_filenames:
    base_name, ext = os.path.splitext(filename)
    if ext in {'.bmp', '.xml'}:
        parts = base_name.split('_')
        if len(parts) == 4:
            patient_id = parts[1]
            record_files[base_name].append(filename)
            # Count unique records (one per base name)
            if len(record_files[base_name]) == 2:  # Both image and annotation
                patient_record_count[patient_id] += 1
```

```python
# Calculate the total number of records and the number of records to move
total_records = len(record_files)
num_records_to_move = int(0.10 * total_records)
print(total_records)
print(num_records_to_move)
```

```python
# Sort patients by their ID and select the patients to move to Directory A and Directory B
sorted_patients = sorted(patient_record_count.keys(), key=lambda id: int(id))
patients_to_move_A = []
patients_to_move_B = []
cumulative_records_A = 0
cumulative_records_B = 0
```

```python
for patient_id in sorted_patients:
    if cumulative_records_A < num_records_to_move:
        patients_to_move_A.append(patient_id)
        cumulative_records_A += patient_record_count[patient_id]
    elif cumulative_records_B < num_records_to_move:
        patients_to_move_B.append(patient_id)
        cumulative_records_B += patient_record_count[patient_id]
    else:
        break
```

```python
# Move the records of the selected patients to the respective directories
for base_name, files in record_files.items():
    patient_id = base_name.split('_')[1]
    if patient_id in patients_to_move_A:
        for file in files:
            if file.endswith('.bmp'):
                source_file = os.path.join(source_directory, file)
                destination_file = os.path.join(image_directory_T, file)
                shutil.move(source_file, destination_file)
                print(f'Moved image to Directory A: {file}')
            elif file.endswith('.xml'):
                source_file = os.path.join(source_directory, file)
                destination_file = os.path.join(label_directory_T, file)
                shutil.move(source_file, destination_file)
                print(f'Moved annotation to Labels for A: {file}')
    elif patient_id in patients_to_move_B:
        for file in files:
            if file.endswith('.bmp'):
                source_file = os.path.join(source_directory, file)
                destination_file = os.path.join(image_directory_V, file)
                shutil.move(source_file, destination_file)
                print(f'Moved image to Directory B: {file}')
            elif file.endswith('.xml'):
                source_file = os.path.join(source_directory, file)
                destination_file = os.path.join(label_directory_V, file)
                shutil.move(source_file, destination_file)
                print(f'Moved annotation to Labels for B: {file}')

print(f"Moved records of patients to Directory A and Labels for A: {patients_to_move_A}")
print(f"Moved records of patients to Directory B and Labels for B: {patients_to_move_B}")
```

# Moving the Remains (The 80%)

```python
images_train_directory = '/content/drive/MyDrive/dataset/images/train'
labels_train_directory = '/content/drive/MyDrive/dataset/labels/train'
```

```python
all_filenames = os.listdir(source_directory)
```

```python
# Separate image files from annotation files
image_filenames = [f for f in all_filenames if f.endswith('.bmp')]
annotation_filenames = [f for f in all_filenames if f.endswith('.xml')]
```

```python
# Create a set of base names for image files
image_base_names = set(os.path.splitext(f)[0] for f in image_filenames)
```

```python
# Move remaining images and their corresponding annotations
for base_name in image_base_names:
    # Move the image file
    image_file = base_name + '.bmp'
    image_source = os.path.join(source_directory, image_file)
    image_dest = os.path.join(images_train_directory, image_file)
    if os.path.exists(image_source):
        shutil.move(image_source, image_dest)

    # Move the annotation file
    annotation_file = base_name + '.xml'
    annotation_source = os.path.join(source_directory, annotation_file)
    annotation_dest = os.path.join(labels_train_directory, annotation_file)
    if os.path.exists(annotation_source):
        shutil.move(annotation_source, annotation_dest)

print("Remaining images and annotations have been moved.")
```
